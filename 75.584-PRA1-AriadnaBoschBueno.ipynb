Ara unirem les columnes coincidents: month, division i line

```python
# Apliquem un "outer merge" per tal que faci la unió completa de les files d'ambdós dataframes, amb valors faltants on no hi ha coincidències.
df_unit = pd.merge(df1, df2, on=['month', 'division', 'line'], how='outer')
df_unit.head()
```

Tenint en compte el nombre de files i columnes dels dataframes originals per separat i del dataframe fusionat, podem dir que la fusió s'ha dut a terme com esperàvem.

Ara, de la columna 'month', que conté el mes i l'any, extraurem l'any per crear una nova columna.

```python
# Primer convertim la columna 'month' a datetime
df_unit['month'] = pd.to_datetime(df_unit['month'], format='%Y-%m')

# Creem la nova columna 'year' extreient l'any de la columna 'month'
df_unit['year'] = df_unit['month'].dt.year

# Mostrem les primeres files per verificar els canvis
df_unit.head()
```

```python
# Calculem el percentatge de valors faltants respecte al de files

percentatge_valors_faltants = (df_unit.isnull().sum() / len(df_unit)) * 100
print(percentatge_valors_faltants)
```

Com podem veure, el nombre de valors faltants és elevat. Per tractar aquestes dades, haurem de decidir quin mètode apliquem.

Tenint en compte que tenim la data completa a la variable 'month', podem utilitzar aquesta informació per completar els valors faltants en la variable 'day_type', indicant si és dia laborable o cap de setmana.

```python
# Importem la llibreria datetime
from datetime import datetime

# Creem una funció per determinar si una data és dia laborable o cap de setmana
def determinar_day_type(data):
    # Retorna 1 si és dia laborable (de dilluns a divendres), 2 si és cap de setmana (dissabte o diumenge)
    if data.weekday() < 5:  # weekday() retorna 0 per dilluns, 1 per dimarts, ..., 6 per diumenge
        return 1  # Dia laborable
    else:
        return 2  # Cap de setmana

# Apliquem la funció a la columna 'month' per crear la nova columna 'day_type'
df_unit['day_type'] = df_unit['month'].apply(determinar_day_type)

# Mostrem les primeres files per verificar el canvi
df_unit.head()
```

Degut a que el nostre objectiu és valorar com el tipus d'incidència (variable 'category') afecta a l'experiència de l'usuari, aquesta variable la conservarem. Per als seus valors faltants afegirem un nou tipus de valor anomenat 'unknown'.

```python
# Substituim els valors faltants en la columna 'category' amb 'unknown'
df_unit['category'] = df_unit['category'].fillna('unknown')

# Mostrem les primeres files per verificar el canvi
df_unit.head()
```

Ara farem el mateix amb els valors faltants de la variable 'period'.

```python
# Substituim els valors faltants en la columna 'period' amb 'unknown'
df_unit['period'] = df_unit['period'].fillna('unknown')
```

Els valors faltants de les variables 'division' i 'line' són molt escassos i decidim eliminar-los.

```python
# Eliminem les files on hi ha valors faltants de les variables 'division' i 'line'
df_unit = df_unit.dropna(subset=['division', 'line'])
```

Ara decidim canviar el nom de la variable 'month', ja que és més correcte si l'anomenem 'date'

```python
# Canviem el nom de la columna 'month' a 'date'
df_unit = df_unit.rename(columns={'month': 'date'})
```

```python
# Calculem el percentatge de valors faltants respecte al de files
percentatge_valors_faltants = (df_unit.isnull().sum() / len(df_unit)) * 100
print(percentatge_valors_faltants)
```

Ara aplicarem un PCA (Principal Component Analysis). Però abans de fer-ho haurem de tractar els valors NaN del nostre dataframe df_unit. Decidim utilitzar un mètode d'imputació per tal de conservar aquests registres i no perdre dades. Els hi imputarem la mitjana de valors de la variable a la qual pertanyen.

```python
from sklearn.impute import SimpleImputer

# Seleccionem només les columnes numèriques per a l'imputació
df_numeric = df_unit.select_dtypes(include=[np.number])

# Imputem els valors NaN amb la mitjana de cada columna
imputer = SimpleImputer(strategy='mean')
df_imputed = imputer.fit_transform(df_numeric)

# Convertim el resultat de la imputació de nou en un dataframe
df_imputed = pd.DataFrame(df_imputed, columns=df_numeric.columns)
```

Ara normalitzarem les dades i mirarem quin és el nombre òptim de components principals a utilitzar.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd

# Normalitzem les dades numèriques imputades
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_imputed)

# Calculem PCA sense restringir el nombre de components
pca = PCA().fit(df_scaled)

# Mostrem el gràfic de la variància explicada acumulada
plt.figure()
plt.plot(pca.explained_variance_ratio_.cumsum())
plt.xlabel('Nombre de Components')
plt.ylabel('Variança acumulada explicada')
plt.show()

# Basant-nos en el gràfic, decidim el nombre de components a utilitzar (amb la variància explicada acumulada fins al 95%)
n_components = sum(pca.explained_variance_ratio_.cumsum() <= 0.95)
print(f"Nombre de components a retenir: {n_components}")

# Ara apliquem PCA amb el nombre de components decidit
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(df_scaled)

# Convertim els components principals en un dataframe per a una visualització més fàcil
df_pca = pd.DataFrame(data=principal_components)
df_pca.columns = ['PC{}'.format(i+1) for i in range(n_components)]
```

Com podem veure, el nombre de components a retenir que expliquen el 95% de la variança és de 4. Llavors, aquest és el nombre de components que utilitzarem.

Ara executem un PCA amb quatre components principals:

```python
# Ara apliquem PCA amb el nombre de components decidit
pca = PCA(n_components=4)
principal_components = pca.fit_transform(df_scaled)

# Convertim els components principals en un dataframe per a una visualització més fàcil
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3', 'PC4'])
```

Ara visualitzem el PCA mitjançant gràfics de dispersió, per parelles de components.

```python
import matplotlib.pyplot as plt

# Creem subgràfics per a cada combinació de dos components
fig, axs = plt.subplots(2, 3, figsize=(15, 10)) # 2 files x 3 columnes

# Primer i segon component
axs[0, 0].scatter(df_pca['PC1'], df_pca['PC2'])
axs[0, 0].set_xlabel('PC1')
axs[0, 0].set_ylabel('PC2')
axs[0, 0].set_title('PC1 i PC2')

# Primer i tercer component
axs[0, 1].scatter(df_pca['PC1'], df_pca['PC3'])
axs[0, 1].set_xlabel('PC1')
axs[0, 1].set_ylabel('PC3')
axs[0, 1].set_title('PC1 i PC3')

# Primer i quart component
axs[0, 2].scatter(df_pca['PC1'], df_pca['PC4'])
axs[0, 2].set_xlabel('PC1')
axs[0, 2].set_ylabel('PC4')
axs[0, 2].set_title('PC1 i PC4')

# Segon i tercer component
axs[1, 0].scatter(df_pca['PC2'], df_pca['PC3'])
axs[1, 0].set_xlabel('PC2')
axs[1, 0].set_ylabel('PC3')
axs[1, 0].set_title('PC2 i PC3')

# Segon i quart component
axs[1, 1].scatter(df_pca['PC2'], df_pca['PC4'])
axs[1, 1].set_xlabel('PC2')
axs[1, 1].set_ylabel('PC4')
axs[1, 1].set_title('PC2 i PC4')

# Tercer i quart component
axs[1, 2].scatter(df_pca['PC3'], df_pca['PC4'])
axs[1, 2].set_xlabel('PC3')
axs[1, 2].set_ylabel('PC4')
axs[1, 2].set_title('PC3 i PC4')

# Ajustem els subgràfics per evitar superposicions
plt.tight_layout()

# Mostrem el gràfic
plt.show()
```

- PC1 i PC2: El primer gràfic mostra una agrupació de punts al llarg de l'eix PC1 i una dispersió considerable al llarg de l'eix PC2, suggerint que PC1 captura una font major de variància, mentre que PC2 captura una font secundària però significativa de variància dins les dades.

- PC1 i PC3: En aquest gràfic, els punts semblen estar més distribuïts al llarg de l'eix PC3 que en el gràfic anterior de PC2, indicant que PC3 pot estar capturant una direcció de variabilitat diferent que no està tan correlacionada amb PC1.

- PC1 i PC4: La dispersió aquí és menys pronunciada que en els gràfics anteriors, la qual cosa podria indicar que PC4 captura menys variància dins de les dades que PC1.

- PC2 i PC3: Aquest gràfic mostra una distribució bastant uniforme de punts sense cap patró de clusterització evident, el que podria indicar que les variàncies que capten PC2 i PC3 són independents entre si.

- PC2 i PC4: Similar al gràfic anterior, no hi ha una agrupació clara de punts, suggerint que la variància que PC4 captura és relativament independent de la variància capturada per PC2.

- PC3 i PC4: Aquí, els punts es distribueixen d'una manera que podria suggerir alguna correlació entre PC3 i PC4, amb una dispersió que indica que ambdós components podrien estar capturant diferents aspectes de la mateixa font de variabilitat.

Ara visualitzarem quina càrrega de variable té cadascun dels 4 components. Això ens servirà per entendre com cada variable original contribueix als components principals que s'han format.
